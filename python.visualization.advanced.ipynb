```
import pandas as pd
df = pd.read_csv('lifeExp.csv')
df.head()
```

```

import seaborn as sns

sns.catplot(x='continent', y='lifeExp', data=df,
            height=4,aspect=1.5,
            kind='bar')
```

```
import seaborn as sns

sns.catplot(x='continent', y='lifeExp', data=df[(df['continent'] !='Europe')],
            height=4,aspect=1.5,
            kind='bar')
```

```
g=sns.lmplot(data=df, x='lifeExp', y='pop', hue='continent',fit_reg=False)
```

```
sns.catplot(x='year', y='lifeExp', data=df,
            height=4,aspect=1.5,
            kind='bar')
```

```
sns.catplot(x='year', y='pop', data=df,
            height=4,aspect=1.5,
            kind='bar')
```

```
sns.catplot(x='year', y='pop', data=df[(df['continent'] =='Asia')],height=4,aspect=1.5,kind='bar')
```

```

sns.catplot(x="year", y="pop", kind="point",data=df[(df['continent'] =='Asia')],hue='country',palette="pastel")
```

```
df_Asia=df[(df['continent'] =='Asia')]
df_Asia=df_Asia[(df_Asia['pop'] >84253330)]
sns.catplot(x="year", y="lifeExp", kind="point",data=df_Asia,hue='country',palette="pastel")
```

```
df_Asia=df[(df['continent'] =='Asia')]
result = df_Asia.pivot(index='country', columns='year', values='lifeExp')
sns.heatmap(result, fmt="g",center=0,linewidths=.5)
```
